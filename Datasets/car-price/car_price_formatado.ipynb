```python
df = df.iloc[:,1:] # Desconsiderar coluna ID
```

```python
df.columns = ['preco', 'imposto', 'fabricante', 'modelo', 'ano_fabricacao',
              'categoria', 'couro_interior', 'tipo_combustivel', 'tamanho_motor_vol',
              'km_rodados','cilindradas', 'tipo_transmissao', 'tracao_rodas', 'qtd_portas', 
              'posicao_volante', 'cor','airbags']
```

```python
df.preco.describe().round(2)
```

```python
df.shape
```

```python
df['ano_fabricacao'] = df['ano_fabricacao'].astype('string')
```

```python
df['couro_interior'].unique()

df['couro_interior'].replace({'Yes':1,'No':0},inplace=True)
```

```python
df.tipo_combustivel.value_counts()
```

```python
df.tamanho_motor_vol.unique()
```

```python
df_tamanho_motor = df.tamanho_motor_vol.str.split(' ',expand=True)
df_tamanho_motor.columns = ['volume_motor','tipo_motor']


df_tamanho_motor.volume_motor = df_tamanho_motor.volume_motor.astype('float')

df_tamanho_motor.tipo_motor.fillna('Comum',inplace=True)

df_tamanho_motor
```

```python
df = pd.concat([df.drop(columns=['tamanho_motor_vol']),df_tamanho_motor],axis=1)
df
```

```python
df.km_rodados = df.km_rodados.str.replace(' km','').astype('int')
```

```python
df.cilindradas = df.cilindradas.astype('int')
```

```python
df.tipo_transmissao.unique()
```

```python
df.tracao_rodas.unique()
```

```python
df.qtd_portas.unique()
```

```python
df.qtd_portas = df.qtd_portas.str.replace('04-May','4_portas').str.replace('02-Mar','2_portas').str.replace('>5','5_ou_mais_portas')
```

```python
df.posicao_volante.unique()
```

```python
df.cor.unique()
```

```python
df.airbags.unique()
```

```python
df.airbags.value_counts()
```

```python
set(sorted(df.fabricante.to_list()))
```

```python
df['imposto']
```

```python
df['imposto'] = df['imposto'].str.replace('-','0').astype('int')
```

```python
df.modelo = df.modelo.str.upper()
```

```python
df
```

## Feature engineering

```python
df['preco_total'] = df.preco + df.imposto
```

```python
df = df.drop(columns=['preco','imposto'])
```

```python
df.preco_total.describe().round(2)
```

```python
df.preco_total.describe().round(2)
```

```python
preco_max_90 = df.preco_total.quantile(.90)
preco_min_90 = df.preco_total.quantile(.10)

print('Max = ',preco_max_90.round(2))

print('Min = ',preco_min_90.round(2))
```

```python
df_final = df.query('@preco_min_90 < preco_total < @preco_max_90')
```
