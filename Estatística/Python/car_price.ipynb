## Probabilidade de um carro custar menos que 10.000$ ?

```python
import scipy.stats as stats
import numpy as np
```

```python
media = df_car['price'].mean()
desvio_padrao = df_car['price'].std()
Z = (10000 - media) / desvio_padrao
```

```python
probabilidade = stats.norm.cdf(Z) * 100
probabilidade
```

## Probabilidade de um carro custar entre 10.000 e 15.000$ ?

```python
media = df_car['price'].mean()
desvio_padrao = df_car['price'].std()

z_inferior = (10000 - media) / desvio_padrao
z_superior = (15000 - media) / desvio_padrao
```

```python
(stats.norm.cdf(z_superior) - stats.norm.cdf(z_inferior)) * 100
```
