Pontos importantes da correlação:

* O coeficiente de correlação mede até que ponto duas variáveis são associados um ao outro.
* Quando valores altos de v1 vão com valores altos de v2, v1 e v2 são associados positivamente.
* Quando valores altos de v1 estão associados a valores baixos de v2, v1 e v2 estão negativamente associados.
* O coeficiente de correlação é uma métrica padronizada para que sempre varie de –1 (correlação negativa perfeita) a +1 (correlação positiva perfeita).
* Um coeficiente de correlação de 0 indica nenhuma correlação, mas esteja ciente de que arranjos aleatórios de dados produzirão valores positivos e negativos para o coeficiente de correlação apenas por acaso.

#### **Coeficiente de correlação**

### 2.1 Amostragem aleatória e viés de amostra

Uma amostra é um subconjunto de dados de um conjunto de dados maior; os estatísticos chamam esse conjunto de dados maior de população.

---
* **Amostra**: Um subconjunto de um conjunto de dados maior(população).

* **População**: O maior conjunto de dados ou ideia de um conjunto de dados.

* **N (n)**: O tamanho da população (amostra).

* **Viés de amostra**: Uma amostra que deturpa a população.

Tipos de amostra

* **Amostragem aleatória**: Desenhar elementos em uma amostra aleatoriamente.

* **Amostragem estratificada**: Divisão da população em estratos e amostragem aleatória de cada um estratos.

* **Amostra aleatória simples**: A amostra que resulta da amostragem aleatória sem estratificar a população. 
---


#### **Amostragem aleatória** 

É um processo no qual cada membro disponível da população que está sendo amostrado tem uma chance igual de ser escolhido para a amostra em cada sorteio.

A amostra resultante é chamada de amostra aleatória simples. 

A amostragem pode ser feita com reposição, na qual as observações são recolocadas na população após cada sorteio para possível nova seleção futura. Ou pode ser feito sem reposição, caso em que as observações, uma vez selecionadas,
ficam indisponíveis para sorteios futuros.

    A qualidade dos dados geralmente é mais importante do que a quantidade de dados ao fazer uma estimativa ou um modelo baseado em uma amostra. A qualidade dos dados na ciência de dados envolve integridade, consistência de formato, limpeza e precisão de pontos de dados individuais. 
    A estatística acrescenta a noção de representatividade.

#### 2.2.1 Regressão à média

A regressão à média refere-se a um fenômeno que envolve medições sucessivas de
uma determinada variável: observações extremas tendem a ser seguidas por outras mais centrais.

Atribuir foco e significado especiais ao valor extremo pode levar a uma forma de viés de seleção.
Os fãs de esportes estão familiarizados com o fenômeno “novato do ano, segundo ano em
queda”. Entre os atletas que iniciam a carreira em uma determinada temporada (classe estreante),
sempre há um que tem um desempenho melhor do que todos os demais.

Geralmente, esse “novato do ano” não se sai tão bem em seu segundo ano. 

Por que não?

Em quase todos os esportes principais, pelo menos naqueles praticados com bola ou disco, há dois
elementos que desempenham um papel no desempenho geral:
    
- Habilidade
- Sorte

A regressão à média é consequência de uma forma particular de viés de seleção.

Quando selecionamos o novato com o melhor desempenho, habilidade e boa sorte provavelmente
estão contribuindo. Em sua próxima temporada, a habilidade ainda estará lá, mas, na maioria dos
casos, a sorte não, então seu desempenho diminuirá - ele regredirá.

     "Por mais que sejam as coisas sejam horríveis ou ruins, sempre se resolverão. Não quer dizer ficarem melhores , apenas quer dizer como se não pudessem ser sempre ruins. Não importa como como ficam horríveis nem ruins, elas voltam sempre ao meio termo, regressão à média. É o equilíbrio entre o bem e o mal."


### 2.3 Sampling Distribution of a Statistic

O termo distribuição amostral de uma estatística refere-se à distribuição de alguma estatística amostral sobre muitas amostras extraídas da
mesma população. 

Grande parte da estatística clássica se preocupa em fazer inferências de amostras
(pequenas) para populações (muito grandes).

---

* **Estatística de amostra**: Uma métrica calculada para uma amostra de dados extraída de uma população maior.

* **Distribuição de dados**: A distribuição de frequência de valores individuais em um conjunto de dados.

* **Distribuição de amostras**: A distribuição de frequência de uma estatística de amostra em muitas amostras ou reamostras.

* **Teorema do limite central**: A tendência da distribuição amostral de assumir uma forma normal à medida que o tamanho
da amostra aumenta.

* **Erro padrão**: A variabilidade (desvio padrão) de uma estatística de amostra em muitas amostras (não confundir com o desvio padrão, que, por si só, se refere à variabilidade de valores de dados individuais).
---

    A distribuição de uma estatística de amostra, como a média, provavelmente será mais regular e em forma de sino do que a distribuição dos próprios dados. Quanto maior a amostra na qual a estatística se baseia, mais isso é verdadeiro. Além disso, quanto maior a amostra, mais estreita é a distribuição da estatística amostral.

#### 2.3.1 Teorema do Limite Central

Este fenômeno é chamado de teorema do limite central. 

Ele diz que as médias extraídas de várias amostras se assemelharão à familiar curva normal em forma de sino (consulte “Distribuição normal”), mesmo que a população de origem não seja normalmente distribuída, desde que o tamanho da amostra seja grande o suficiente e a partida dos dados da normalidade não é muito grande.

#### 2.3.2 Erro padrão

O erro padrão é uma única métrica que resume a variabilidade na distribuição de amostragem
para uma estatística.

À medida que o tamanho da amostra aumenta, o erro padrão diminui (...); A relação entre o erro padrão e o tamanho da amostra às vezes é
chamada de regra da raiz quadrada de n, para reduzir o erro padrão em um fator de 2, o tamanho da amostra deve ser aumentado em um fator de 4.

Na verdade, você não precisa confiar no teorema do limite central para
entender o erro padrão. Considere a seguinte abordagem para medir o erro padrão:

1. Colete um número de novas amostras da população.

2. Para cada nova amostra, calcule a estatística (por exemplo, média).

3. Calcule o desvio padrão das estatísticas calculadas no passo 2; use isso como sua estimativa de erro padrão.

Na prática, essa abordagem de coletar novas amostras para estimar o erro padrão normalmente não é viável (e estatisticamente muito inútil). Felizmente, verifica-se que não é necessário retirar novas amostras; em vez disso, você pode usar reamostragens bootstrap (consulte “O Bootstrap”). 

Nas estatísticas modernas, o bootstrap tornou-se a maneira padrão de estimar o erro padrão. Ele pode ser usado para praticamente qualquer estatística e não depende do teorema do limite central ou outro.

* S = Erro padrão
* N = Tamanho da amostra

### 2.4 The Bootstrap

Uma maneira fácil e eficaz de estimar a distribuição amostral de uma estatística ou de parâmetros
de modelo é extrair amostras adicionais, com reposição, da própria amostra e recalcular a
estatística ou modelo para cada reamostragem.

Conceitualmente, você pode imaginar o bootstrap replicando a amostra original milhares ou milhões de
vezes para que você tenha uma população hipotética que incorpora todo o conhecimento de sua amostra
original (é apenas maior). Você pode então extrair amostras dessa população hipotética com o objetivo de
estimar uma distribuição amostral.



    "Imagine que você está interessado em estimar uma característica de uma população, como a média ou a mediana. No entanto, você só tem acesso a uma amostra dessa população, e não tem os recursos para coletar mais dados. 
    
    O que você pode fazer nesse caso?

    É aí que entra o bootstrap. O bootstrap é um método estatístico que PERMITE ESTIMAR A INCERTEZA ASSOCIADA A UMA ESTIMATIVA, utilizando apenas a amostra disponível.

    A ideia básica do bootstrap é simular a variabilidade da amostra através da reamostragem. Você cria várias amostras "fictícias" da mesma população, chamadas de amostras bootstrap, fazendo seleções aleatórias com reposição da amostra original. A seleção com reposição significa que você pode escolher o mesmo elemento da amostra original várias vezes ou até mesmo não selecioná-lo em uma amostra bootstrap.

    Após criar várias amostras bootstrap, você pode calcular a estimativa da característica de interesse em cada uma delas. Por exemplo, se você estiver interessado na média, calcularia a média em cada amostra bootstrap.

    A grande sacada é que, ao repetir esse processo de reamostragem várias vezes (geralmente milhares de vezes), você obtém uma distribuição de estimativas da característica de interesse. Essa distribuição é chamada de distribuição bootstrap.

    A partir dessa distribuição, é possível calcular intervalos de confiança para a estimativa. Por exemplo, um intervalo de confiança de 95% pode ser obtido selecionando os percentis 2.5% e 97.5% da distribuição bootstrap.

    Em resumo, o bootstrap é uma ferramenta poderosa na estatística, pois permite estimar a incerteza associada a uma estimativa sem a necessidade de coletar mais dados. Ele utiliza a ideia de reamostragem para criar várias amostras fictícias da mesma população e, a partir dessas amostras, é possível obter uma distribuição de estimativas e calcular intervalos de confiança. Isso ajuda a ter uma noção da variabilidade dos resultados e da confiabilidade da estimativa obtida a partir da amostra original."

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados de exemplo
seed = np.random.seed(seed=12)
dados = np.random.randint(low=0,high=1000,size=54000)

# Função para realizar o Bootstrap
def bootstrap(dados, n_iteracoes, estatistica):
    estatisticas_bootstrap = []
    n = len(dados)
    
    for _ in range(n_iteracoes):
        amostra = np.random.choice(dados, size=n, replace=True)
        estatistica_bootstrap = estatistica(amostra)
        estatisticas_bootstrap.append(estatistica_bootstrap)
    
    return estatisticas_bootstrap

# Executar o Bootstrap para estimar a média
estatisticas_bootstrap = bootstrap(dados, n_iteracoes=1000, estatistica=np.mean)
print(f'Mediana real: {np.mean(dados)}\n')

print(f'Mediana Bootstrap: {np.mean(estatisticas_bootstrap)}')
print(f'Desvio Padrão Bootstrap: {np.std(estatisticas_bootstrap)}')

# Histograma das estatísticas bootstrap
plt.hist(estatisticas_bootstrap, bins=30)
plt.xlabel('Média')
plt.ylabel('Frequência')
plt.title('Distribuição Bootstrap da Média')
plt.show()
```

Quanto mais iterações você fizer, mais precisa será a estimativa do erro padrão ou o intervalo de confiança. 

O resultado desse procedimento é um conjunto de inicialização de estatísticas de amostra ou
parâmetros de modelo estimados, que você pode examinar para ver como eles são variáveis.

```python
estatisticas_bootstrap = bootstrap(dados, n_iteracoes=10000, estatistica=np.mean)
print(f'Média real: {np.mean(dados)}\n')

print(f'Média Bootstrap: {np.mean(estatisticas_bootstrap)}')
print(f'Desvio Padrão Bootstrap: {np.std(estatisticas_bootstrap)}')

# Histograma das estatísticas bootstrap
plt.hist(estatisticas_bootstrap, bins=30)
plt.xlabel('Média')
plt.ylabel('Frequência')
plt.title('Distribuição Bootstrap da Média')
plt.show()

estatisticas_bootstrap = bootstrap(dados, n_iteracoes=100000, estatistica=np.mean)
print(f'Média real: {np.mean(dados)}\n')

print(f'Média Bootstrap: {np.mean(estatisticas_bootstrap)}')
print(f'Desvio Padrão Bootstrap: {np.std(estatisticas_bootstrap)}')

# Histograma das estatísticas bootstrap
plt.hist(estatisticas_bootstrap, bins=30)
plt.xlabel('Média')
plt.ylabel('Frequência')
plt.title('Distribuição Bootstrap da Média')
plt.show()
```

#### Exemplo:

Suponha que você tenha uma amostra de dados que representa o tempo de resposta de um sistema. Você deseja estimar a média desse tempo de resposta e também obter um intervalo de confiança para essa estimativa.

```python
import numpy as np

# Amostra de dados (tempo de resposta)
amostra = np.array([10.2, 9.5, 11.1, 10.8, 9.7, 10.6, 9.9, 11.3, 10.4, 10.0])

def bootstrap(dados, n_iteracoes, estatistica):
    estatisticas_bootstrap = []
    n = len(dados)
    
    for _ in range(n_iteracoes):
        amostra = np.random.choice(dados, size=n, replace=True)
        estatistica_bootstrap = estatistica(amostra)
        estatisticas_bootstrap.append(estatistica_bootstrap)
    
    return estatisticas_bootstrap

# Executar o Bootstrap para estimar a média
estatisticas_bootstrap = bootstrap(amostra, n_iteracoes=1000, estatistica=np.mean)
print(f'Média real: {np.mean(amostra)}\n')

print(f'Média Bootstrap: {np.mean(estatisticas_bootstrap)}')
print(f'Desvio Padrão Bootstrap: {np.std(estatisticas_bootstrap)}')

# Histograma das estatísticas bootstrap
plt.hist(estatisticas_bootstrap, bins=30)
plt.xlabel('Média')
plt.ylabel('Frequência')
plt.title('Distribuição Bootstrap da Média')
plt.show()


# Calcula o intervalo de confiança de 95% usando os percentis
intervalo_confianca = np.percentile(estatisticas_bootstrap, [2.5, 97.5])

print("Intervalo de confiança (95%):", intervalo_confianca)
```

### 2.5 Confidence Intervals

---
* **Nível de confiança**: A porcentagem de intervalos de confiança, construídos da mesma forma a partir da mesma população, que se espera conter a estatística de interesse.
    
* **Pontos finais do intervalo**: A parte superior e inferior do intervalo de confiança.
---

Existe uma aversão humana natural à incerteza; as pessoas (especialmente os especialistas) raramente dizem “não sei”. 

Analistas e gerentes, embora reconheçam a incerteza, depositam fé indevida em uma estimativa quando ela é apresentada como um único número (uma estimativa pontual). 

Apresentar uma estimativa não como um número único, mas como um intervalo é uma forma de neutralizar essa tendência.

    Claro, o que realmente nos interessa quando temos um resultado de amostra é “qual é a probabilidade de que o valor verdadeiro esteja dentro de um certo intervalo?” Esta não é realmente a pergunta que um intervalo de confiança responde, mas acaba sendo como a maioria das pessoas interpreta a resposta.
    
    A questão de probabilidade associada a um intervalo de confiança começa com a frase “Dado um procedimento de amostragem e uma população, qual é a probabilidade de que…” Para ir na direção oposta, “Dado um resultado de amostra, qual é a probabilidade de que (algo seja verdade sobre a população)”, envolve cálculos mais complexos e imponderáveis mais profundos.

IDEIAS-CHAVE
* Intervalos de confiança são a maneira típica de apresentar estimativas como um intervalo de intervalo.
* Quanto mais dados você tiver, menos variável será uma estimativa de amostra.
* Quanto menor o nível de confiança que você pode tolerar, mais estreito será o intervalo de confiança.
* O bootstrap é uma maneira eficaz de construir intervalos de confiança.

### 2.8 Student’s t-Distribution

A distribuição t é uma distribuição de formato normal, mas um pouco mais espessa e mais longa nas caudas. É amplamente utilizado para representar distribuições de estatísticas de amostra.

As distribuições de médias amostrais geralmente têm a forma de uma distribuição t, e há uma família de distribuições t que diferem dependendo do tamanho da amostra. Quanto maior a amostra, mais normal se torna a distribuição t.

A distribuição t é frequentemente chamada de t de Student porque foi publicada em 1908 na Biometrika por WS Gossett sob o nome de “Student”. 
O empregador de Gossett, a cervejaria Guinness, não queria que os concorrentes soubessem que estava usando métodos estatísticos, então insistiu que Gossett não usasse seu nome no artigo.

Gossett queria responder à pergunta “Qual é a distribuição amostral da média de uma amostra, extraída de uma população maior?”

Ele começou com um experimento de reamostragem extraindo amostras aleatórias de 4 de um conjunto de dados de 3.000 medidas de altura de criminosos e comprimentos do dedo médio esquerdo. (Sendo esta a era da eugenia, havia muito interesse em dados sobre criminosos e em descobrir correlações entre tendências criminosas e atributos físicos ou psicológicos.) 

Ele traçou os resultados padronizados (os escores z) no eixo x e a frequência no eixo y.

Separadamente, ele derivou uma função, agora conhecida como t de Student, e ajustou essa função aos resultados da amostra, plotando a comparação.

Várias estatísticas diferentes podem ser comparadas, após a padronização, à distribuição t, para estimar os intervalos de confiança à luz da variação da amostragem.

Considere uma amostra de tamanho n para a qual a média amostral foi calculada. 
Se s é o desvio padrão da amostra, um intervalo de confiança de 90% em torno da média da amostra é dado por:



onde é o valor da estatística t, com (n – 1) graus de liberdade (consulte “Graus de liberdade”), que “corta” 5% da distribuição t em cada extremidade. 

A distribuição t tem sido usada como referência para a distribuição de uma média amostral, a diferença entre duas médias amostrais,
parâmetros de regressão e outras estatísticas.


Se o poder de computação estivesse amplamente disponível em 1908, as estatísticas, sem dúvida, teriam se baseado muito mais fortemente em métodos de reamostragem computacionalmente intensivos desde o início. Na falta de computadores, os estatísticos recorreram à matemática e a funções como a distribuição t para aproximar as distribuições amostrais.


O poder do computador permitiu experimentos práticos de reamostragem na década de 1980, mas até então, o uso da distribuição t e distribuições semelhantes tornou-se profundamente incorporado em livros didáticos e software.


A precisão da distribuição t em representar o comportamento de uma estatística de amostra exige que a distribuição dessa estatística para essa amostra seja moldada como uma distribuição normal. Acontece que as estatísticas amostrais geralmente são normalmente distribuídas, mesmo quando os dados populacionais subjacentes não são (um fato que levou à aplicação generalizada da distribuição t).

### 2.9 Binomial Distribution

Os resultados sim/não (binomiais) estão no centro da análise, pois geralmente são a culminação de uma decisão ou outro processo; compre/não compre, clique/não clique, sobreviva/morra e assim por diante.

Central para entender a distribuição binomial é a ideia de um conjunto de tentativas, cada tentativa tendo dois resultados possíveis com probabilidades definidas.

Por exemplo, lançar uma moeda 10 vezes é um experimento binomial com 10 tentativas, cada tentativa tendo dois resultados possíveis (cara ou coroa). Esses resultados sim/não ou 0/1 são denominados resultados binários e não precisam ter probabilidades de 50/50. Quaisquer probabilidades que somam 1,0 são possíveis. 

É convencional em estatística denominar o resultado “1” como o resultado de sucesso ; também é prática comum atribuir “1” ao resultado mais raro.

O uso do termo sucesso não implica que o resultado seja desejável ou benéfico, mas tende a indicar o resultado de interesse. Por exemplo, inadimplência de empréstimos ou transações fraudulentas são eventos relativamente incomuns que podemos estar interessados em prever, por isso são denominados “1s” ou “sucessos”.

A distribuição binomial é a distribuição de frequência do número de sucessos (x) em um determinado número de tentativas (n) com probabilidade especificada (p) de sucesso em cada tentativa. 

Existe uma família de distribuições binomiais, dependendo dos valores de x, n e p. A distribuição binomial responderia a uma pergunta como:
Se a probabilidade de um clique converter em uma venda for 0,02, qual é a probabilidade de observar 0 vendas em 200 cliques?

```python
from scipy.stats import binom

x = 0
n = 200
p = 0.02


print(f'Probabilidade de: {(binom.pmf(x,n,p)* 100).round(2)} %')
```

Com um número suficientemente grande de tentativas (particularmente quando p está próximo de 0,50), a distribuição binomial é virtualmente indistinguível da distribuição normal. Na verdade, o cálculo de probabilidades binomiais com grandes tamanhos de amostra é computacionalmente exigente, e a maioria dos procedimentos estatísticos usa a distribuição normal, com média e variância, como uma aproximação.

IDEIAS-CHAVE

- Os resultados binomiais são importantes para modelar, pois representam, entre outras coisas, decisões fundamentais (comprar ou não comprar, clicar ou não clicar, sobreviver ou morrer, etc.). 
- Uma tentativa binomial é um experimento com dois resultados possíveis: um com probabilidade p e outro com probabilidade 1 – p.
- Com n grande e desde que p não esteja muito próximo de 0 ou 1, a distribuição binomial pode ser aproximada pela distribuição normal.
